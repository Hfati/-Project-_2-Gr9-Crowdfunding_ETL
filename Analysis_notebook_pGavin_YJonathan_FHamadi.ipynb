```python
import sqlalchemy
from sqlalchemy import create_engine, inspect, text

# data
import pandas as pd
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
from sklearn import datasets
```

```python
connection_string = f"postgresql+psycopg2://{SQL_USERNAME}:{SQL_PASSWORD}@{SQL_IP}:{PORT}/{DATABASE}"
engine = create_engine(connection_string)
```

```python
SQL_USERNAME = 'postgres'
SQL_PASSWORD = 'postgres'
SQL_IP = 'localhost'
PORT = 5432
DATABASE = 'project_2_db'
```

```python
query = """
    SELECT 
        outcome,
        campaign.category_id,
        category
    FROM
        campaign
    JOIN
        category ON category.category_id = campaign.category_id
    WHERE
        category = 'food'

"""

food_df = pd.read_sql_query(query, engine)
food_df.head()
```

```python
food_data_df = food_df.outcome.value_counts()
```

```python
labels = ['successful','failed','canceled']
colors = ['green','red','yellow']
plt.figure(figsize=(8, 6))

plt.pie(food_data_df, labels=labels, colors=colors,shadow=True)

plt.title('Food Campaigns Success Rate\n')
plt.axis('equal')

plt.show()
```
